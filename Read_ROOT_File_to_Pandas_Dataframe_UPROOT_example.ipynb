# Reading example ROOT file in pandas Dataframe

In this example I use library [uproot](https://masonproffitt.github.io/uproot-tutorial/). More typical way to do it would be use of [JupyROOT package](https://git.ustc.edu.cn/cxwx/root/-/blob/v6-26-02/bindings/jupyroot/README.md) and it's ROOT libary (which is done in another example), but 'uproot' allows one read and write files in the .root format without actually requiring or running the ROOT software at all 

```python
filename = "/home/victoria/Projects_1/AI_onto_all/Files_example_KCDC/Example_ROOT/KASCADE_HighEnergyData_runs_0877-7417_ROOT/events.root"
with uproot.open(filename) as f:
    tree = f["lopes"]
    df = tree.arrays(library="pd")
```
